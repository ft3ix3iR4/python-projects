```python
def invert_colors(grid):
    out_grid = grid.copy()
    for row in range(grid.height):
        for col in range(grid.width):
            out_grid[row, col].red = 255 - grid[row, col].red
            out_grid[row, col].green = 255 - grid[row, col].green
            out_grid[row, col].blue = 255 - grid[row, col].blue
    return out_grid
```

```python
def path_color(grid, path, starting_point, color):
    grid = grid.copy()
    directions = {
        'up': (-1, 0),
        'down': (1, 0),
        'right': (0, 1),
        'left': (0, -1)
    }
    pos = starting_point
    grid[pos] = color
    for direction in path:
        move = directions[direction]
        pos = (pos[0] + move[0], pos[1] + move[1])
        grid[pos] = color
    return grid
```

```python
path_color(grid, ['down', 'down', 'right', 'down'], (1, 1), colors.Aqua)
```
